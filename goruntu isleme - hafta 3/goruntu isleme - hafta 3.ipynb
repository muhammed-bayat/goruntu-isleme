```
def my_function_1(my_list = [2, 4, 3, 4, 5, 6, 3, 3, 2, 1]):
  t = 0
  s = 0
  for i in my_list:
    s = s + 1
    t = t + i
  mean = t / s
  
  t = 0
  s = 0
  
  for i in my_list:
    s = s + 1
    t = t + (i - mean) * (i - mean)
  
  var = t / (s - 1)
  
  return mean, var
```

```
import numpy as np
import matplotlib.pyplot as plt
```

```
image = plt.imread("Venice.jpg")
```

```
def my_function_2(image = plt.imread("Venice.jpg")):
  m, n, p = image.shape
  my_histogram = {}
  for i in range(m):
    for j in range(n):
      if image[i, j, 0] in my_histogram.keys():
        my_histogram[image[i, j, 0]] = my_histogram[image[i, j, 0]] + 1
      else:
        my_histogram[image[i, j, 0]] = 1
  return my_histogram
```

```
def my_function_3(my_histogram = my_function_2()):
  x = []
  y = []
  for key in my_histogram.keys():
    x.append(key)
    y.append(my_histogram[key])
  return x,y
```

```
x, y = my_function_3()
plt.bar(x, y)
plt.show()
```
