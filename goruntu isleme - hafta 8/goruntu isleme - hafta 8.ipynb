```
import numpy as np
import matplotlib.pyplot as plt
```

```
image_1 = plt.imread('bird.jpg')
image_1.shape
```

```
plt.imshow(image_1)
plt.show()
```

```
my_histogram_rgb = {}
m, n, p = image_1.shape
for i in range(m):
  for j in range(n):
    s = (image_1[i, j, 0])
    if(0, s) in my_histogram_rgb.keys():
      my_histogram_rgb[(0, s)] = my_histogram_rgb[(0, s)] + 1
    else:
      my_histogram_rgb[(0, s)] = 1
my_histogram_rgb
```

```
for i in range(m):
  for j in range(n):
    s = (image_1[i, j, 1])
    if(1, s) in my_histogram_rgb.keys():
      my_histogram_rgb[(1, s)] = my_histogram_rgb[(1, s)] + 1
    else:
      my_histogram_rgb[(1, s)] = 1
my_histogram_rgb
```

```
for i in range(m):
  for j in range(n):
    s = (image_1[i, j, 2])
    if(2, s) in my_histogram_rgb.keys():
      my_histogram_rgb[(2, s)] = my_histogram_rgb[(2, s)] + 1
    else:
      my_histogram_rgb[(2, s)] = 1
my_histogram_rgb
```

```
my_histogram = {}
m, n, p = image_1.shape
for i in range(m):
  for j in range(n):
    s = (image_1[i, j, 0], image_1[i, j, 1], image_1[i, j, 2])
    if s in my_histogram.keys():
      my_histogram[s] = my_histogram[s] + 1
    else:
      my_histogram[s] = 1
my_histogram
```
