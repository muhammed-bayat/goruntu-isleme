```
import math
def my_pdf_1(x, mu = 0.0, sigma = np.finfo(float).eps):
  x = float(x - mu) / sigma
  return math.exp(-x * x / 2.0) / math.sqrt(2.0 * math.pi) / np.finfo(float).eps
my_pdf_1(10, 1, 3)
```

```
def get_my_mean_and_std(k = 0, l = 0, m = 10000):
    s = 0
    t = 0
    for i in range(m):
        if(train_data[i, 0] == k):
            s = s + 1
            t = t + train_data[i, l+1]
    mean_1 = t / s

    s, t = 0, 0
    for i in range(m):
        if(train_data[i, 0] == k):
            s = s + 1
            diff_1 = train_data[i, l + 1] - mean_1
            t = t + diff_1 * diff_1
    std_1 = np.sqrt(t / (s - 1))

    print(mean_1, std_1)
    return mean_1, std_1
```
