Eklenen resim yukarıdaki şekilde görüntülenir.

```
new_image = np.zeros((m, n), dtype=int)
new_image
```

Return a new array of given shape and type, **filled with zeros**.

```
for i in range(m):
  for j in range(n):
    s = (image_1[i, j, 0] + image_1[i, j, 1] + image_1[i, j, 2]) / 3
    new_image[i, j] = s
```

```
plt.imshow(new_image, 'gray')
plt.show()
```
