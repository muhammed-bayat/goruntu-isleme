```
def my_function_4(image):
  m, n, p = image.shape
  new_m = int(m / 2)
  new_n = int(n / 2)
  new_image = np.zeros((new_m, new_n), dtype = int)
  for m in range(new_m):
    for n in range(new_n):
      s = (image[m * 2, n * 2, 0] + image[m * 2, n * 2, 1] + image[m * 2, n * 2, 2]) / 3
      new_image[m, n] = int(s)
  return new_image
```

```
def rescale_image(image):
  m, n, p = image.shape
  new_m = int(m / 2)
  new_n = int(n / 2)

  new_image = np.zeros((new_m, new_n, 3), dtype = int)
  for m in range(new_m):
    for n in range(new_n):
      new_image[m, n, 0] = int(image[m * 2, n * 2, 0])
      new_image[m, n, 1] = int(image[m * 2, n * 2, 1]) 
      new_image[m, n, 2] = int(image[m * 2, n * 2, 2])
  return new_image
```
